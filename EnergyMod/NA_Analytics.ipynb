# INTRODUCTION

This notebook seeks to analyse and provide insight into data on communities with no access to electricity in Ghana.

```python
#Importing Libraries for analysis

import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
#extracting data from csv file

df = pd.read_csv("NoAccess.csv")
df.info()
```

```python
df.head()
```

```python
fig = px.scatter_mapbox(
    df, #current dataframe
    lat = "lon",
    lon = "lat",
    center={"lat": 7.94, "lon": -1.023},
    width = 600,
    height = 600,
    hover_data=["Community"],
)

fig.update_layout(mapbox_style = "open-street-map")

fig.show()
```
